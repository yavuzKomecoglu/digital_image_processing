# Görüntü nedir ve bilgisayarda nasıl saklanır

Kavramsal olarak, en basit haliyle bir görüntü (**tek kanallı**; örneğin ikili veya mono-krom, gri tonlamalı veya siyah beyaz görüntüler), koordinat çiftini bir tam sayıya eşleyen iki boyutlu bir işlevdir **f(x, y)**. (x,y koordinatındaki tam sayı renk değeri) 
Her noktaya bir **piksel** denir. Bir görüntüde birden fazla kanal olabilir (örneğin, bir rengin üç kanal kullanarak (kırmızı, yeşil ve mavi) gösterilebileceği renkli RGB görüntüleri). Renkli bir RGB görüntü için, **(x, y)** koordinatındaki her bir piksel, üç-tuple **(rx, y, gx, y, bx, y)** ile temsil edilebilir.

Bilgisayarda işleyebilmek için **f(x, y)** görüntüsünün hem uzamsal hem de genlik olarak dijitalleştirilmesi gerekir. Mekansal koordinatların (x, y) sayısallaştırılmasına görüntü örnekleme adı verilir. Genlik sayısallaştırmaya gri seviye nicelemesi denir. Bir bilgisayarda, bir kanala karşılık gelen bir piksel değeri genellikle (0-255) arasında bir tamsayı değeri veya (0-1) arasında bir kayan nokta değeri olarak temsil edilir. Bir görüntü bir dosya olarak saklanır ve birçok farklı türde dosya (biçim) olabilir. Her dosya genellikle bazı meta verilere ve çok boyutlu diziler olarak çıkarılabilen bazı verilere sahiptir (örneğin, ikili veya gri seviye görüntüler için 2 boyutlu diziler ve RGB ve YUV renkli görüntüler için 3D diziler). Aşağıdaki şekil, resim verilerinin farklı resim türleri için matrisler olarak nasıl depolandığını gösterir. Gösterildiği gibi, gri tonlamalı bir görüntü için, **genişlik x yüksekliğe** sahip bir matris (2-D dizisi) görüntünün depolanması için yeterliyken, bir RGB görüntü için **genişlik x yükseklik x 3** genişliğinde bir boyutun 3-D dizisini gerektirir:

## Görüntü işleme nedir?

Görüntü işleme, bilgisayardaki algoritmalar ve kodlar kullanılarak görüntülerin otomatik olarak işlenmesi, manipülasyonu, analizi ve yorumlanması anlamına gelir. **Televizyon, fotoğrafçılık, robotik, uzaktan algılama, tıbbi teşhis ve endüstriyel muayene gibi bilim ve teknolojide** birçok disiplin ve alanda uygulamaları vardır. Günlük hayatımızda alıştığımız ve her gün tonlarca görüntü yüklediğimiz Facebook ve Instagram gibi sosyal paylaşım siteleri, görüntüleri işlemek için birçok görüntü işleme algoritması kullanması/yenilemesi gereken sektörlerin tipik örnekleridir. 

Bu kitapta, bir görüntüyü işlemek için birkaç Python paketi kullanacağız. Öncelikle, klasik görüntü işleme için bir sürü kütüphaneyi kullanacağız: görüntü verilerini çıkarmaktan, kütüphane işlevlerini kullanarak bazı algoritmalarla verileri dönüştürmekten, önceden işlemek, geliştirmek, geri yüklemek, temsil etmek (tanımlayıcılarla), bölümlemek, sınıflandırmak ve Verileri daha iyi analiz etmek, anlamak ve yorumlamak için tespit etmek ve tanımak (nesneler). Daha sonra, son birkaç yılda çok popüler olan bir teknoloji olan derin öğrenmeye dayalı görüntü işleme yapmak için başka kütüphaneler kullanacağız.
