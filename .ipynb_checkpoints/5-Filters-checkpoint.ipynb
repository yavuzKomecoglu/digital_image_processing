## Median(Ortanca Deer) Filtresi

![alt text](assets/5-median_filter.gif)

### Tuz-buz grltsnden temizleme
Bir pencere boyutu belirlenir ve resim zerinde gezdirilen pencere ierisindeki piksel deerleri sralanarak ortadaki deer alnr.

```python
img_noisy_array = np.array(img_noisy)
new_img_array = np.array(img_noisy)
```

## 3x3 Filtre

```python
def getMedian(li):
    li.sort()
    return li[int(len(li)/2)]
```

```python
############### 3x3 window ###############
for i in range(1, img_noisy_array.shape[0] - 1):
    for j in range(1, img_noisy_array.shape[1] - 1):
        win = []
        for x in range(i - 1, i + 2):
            for y in range(j - 1, j + 2):
                win.append(img_noisy_array[x][y])

        new_img_array[i][j] = getMedian(win)

imshow(new_img_array)
```

## 5x5 Filtre

```python
############### 5x5 window ###############
img_noisy_array = np.array(img_noisy)
new_img_array = np.array(img_noisy)

for i in range(1, img_noisy_array.shape[0] - 2):
    for j in range(1, img_noisy_array.shape[1] - 2):
        win = []
        for x in range(i - 2, i + 3):
            for y in range(j - 2, j + 3):
                win.append(img_noisy_array[x][y])
        #sort the values
        win.sort()
        
        new_img_array[i][j] = getMedian(win)

imshow(new_img_array)
```

## Mean (Ortalama) Filtresi
Resim zerinde gezdirilen pencere ierisindeki piksel deerlerinin ortalamas alnr.

```python
############### 5x5 window ###############
img_noisy_array = np.array(img_noisy)
new_img_array = np.array(img_noisy)


for i in range(1, img_noisy_array.shape[0] - 2):
    for j in range(1, img_noisy_array.shape[1] - 2):
        win = []
        for x in range(i - 2, i + 3):
            for y in range(j - 2, j + 3):
                win.append(img_noisy_array[x][y])
                
        new_img_array[i][j] = (np.sum(win))/(5*5)

imshow(new_img_array)
```

## Kenar nedir

Piksel younluklarnda ar farkllklar bulunan alan genellikle bir nesnenin kenarn gsterir.

### Kenar eitleri

Bunlar, grntlerde bulunan ana ideal kenar tipleridir. Bir grntdeki grnt younluunun deime ekli, o konumda mevcut olan kenar tipini belirler. ekil 1, piksellerin x ya da y dorultusundaki mesafelerine kar olan grnt younluunu temsil etmek iin bir izgi kullanarak bu farkllklar gstermeye yardmc olur.

- a. Adm Kenar (Step Edge) - Grntnn younluunun derhal bir mesafeden artt yerde.
- b. Rampa Kenar (Ramp Edge) - Grntnn younluunun belli bir mesafeye kademeli olarak artt yer.
- c. Tavan Kenar (Roof Edge) - Grntnn younluunun geici olarak belirli bir mesafeye yayld yer.

## Kenar Tespiti: Sobel Filtresi

```python
path = "images/jet.jpg"
img = Image.open(path)

imshow(img)
```

Sobel, bir grntnn kenarlarn alglamak iin ok yaygn bir operatrdr, bu bir grntnn trevine bir yaklamdr. Y ve x ynlerinde ayrdr. Burada, her x ve y yn iin bir tane olmak zere bir ekirdek 3 * 3 matrisi kullanyoruz. X yn iin degrade, solda eksi saylar, sanda da pozitif saylar ve merkez pikselleri koruruz. Satr pikselleri.

![alt text](assets/5-sobel_con_kernels.png)


```python
img_array = np.array(img)

#define horizontal and Vertical sobel kernels
Gx = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
Gy = np.array([[-1, -2, -1],[0, 0, 0],[1, 2, 1]])
```

### ekirdek evriim fonksiyonunu tanmlamak

Evriim ilevi, ayn boyutta nc bir say dizisini retmek iin genellikle farkl boyutlardaki iki say dizisini, burada gri skala grntsn ve sobel / prewitt ekirdeklerini bir araya getirme yntemidir. Evriim, uygulanan ekirdein grntnn tm piksellerinin zerine kaydrlmasyla gerekletirilir.
A, kaynak grntdr ve * evriim ilemini gsterir.


![alt text](assets/5-math_gx.svg)

![alt text](assets/5-math_gy.svg)


```python
#define kernal convolution function
# with image X and filter F
def convolve(X, F):
    # height and width of the image
    X_height = X.shape[0]
    X_width = X.shape[1]
    
    # height and width of the filter
    F_height = F.shape[0]
    F_width = F.shape[1]
    
    H = (F_height - 1) // 2
    W = (F_width - 1) // 2
    
    #output numpy matrix with height and width
    out = np.zeros((X_height, X_width))
    #iterate over all the pixel of image X
    for i in np.arange(H, X_height-H):
        for j in np.arange(W, X_width-W):
            sum = 0
            #iterate over the filter
            for k in np.arange(-H, H+1):
                for l in np.arange(-W, W+1):
                    #get the corresponding value from image and filter
                    a = X[i+k, j+l]
                    w = F[H+k, W+l]
                    sum += (w * a)
            out[i,j] = sum
    #return convolution  
    return out
```

### Gradyan bykl

Her yndeki gradyan bileeni daha sonra her noktada gradyann mutlak bykln ve bu gradyann ynn bulmak iin bir araya getirilir.

```python
#normalizing the vectors
sob_x = convolve(img_array, Gx) / 8.0
sob_y = convolve(img_array, Gy) / 8.0


#calculate the gradient magnitude of vectors
sob_out = np.sqrt(np.power(sob_x, 2) + np.power(sob_y, 2))
# mapping values from 0 to 255
sob_out = (sob_out / np.max(sob_out)) * 255


imshow(sob_out)
```
